```python
import gradio as gr
from transformers import pipeline


summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Define the summarization function
def summarize_text(text):
    summary = summarizer(text, max_length=130, min_length=30, do_sample=False)
    return summary[0]['summary_text']

# Create Gradio interface
iface = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(lines=10, placeholder="Enter text to summarize here..."),
    outputs="text",
    title="Text Summarization",
    description="Summarize any text using the Facebook BART-large-CNN model.",
)

# Launch the interface
iface.launch()
```

```python
import gradio as gr
from transformers import pipeline
import re

# Initialize the summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Function to extract and summarize article details
def summarize_article(text):
    # Extract metadata (Publisher, Author, Date) using simple patterns
    publisher = re.search(r'Publisher:\s*(.*)', text)
    author = re.search(r'Author:\s*(.*)', text)
    date = re.search(r'Date:\s*(.*)', text)

    # Handle missing information
    publisher = publisher.group(1).strip() if publisher else "Unknown"
    author = author.group(1).strip() if author else "Unknown"
    date = date.group(1).strip() if date else "Unknown"

    # Remove metadata lines to get the article content
    content = re.sub(r'(Publisher:.*|Author:.*|Date:.*)', '', text, flags=re.IGNORECASE)

    # Summarize the content
    summary = summarizer(content.strip(), max_length=130, min_length=30, do_sample=False)[0]['summary_text']

    # Return structured output
    return f"**Summary:** {summary}\n\n**Publisher:** {publisher}\n**Author:** {author}\n**Date:** {date}"

# Gradio interface setup
iface = gr.Interface(
    fn=summarize_article,
    inputs=gr.Textbox(lines=20, placeholder="Enter article with Publisher, Author, Date..."),
    outputs="text",
    title="Article Summarizer with Metadata",
    description="Summarize an article and extract key details like Publisher, Author, and Date."
)

# Launch the Gradio interface
iface.launch()
```
